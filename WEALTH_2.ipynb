```
#I am not sure which year data we have for acs so I randomly chose year to
#be 2018 it can be changed and run again to get the results corresponding to
#the year
YEAR = '2018'
```

```
#also making sure that no entry in the total/total_population/totlaAge25plus
#is empty
assert race_ds['total_population'].isnull().sum() == 0
assert age_ds['total'].isnull().sum() == 0
assert education_ds['totalAge25plus'].isnull().sum() == 0
```

```
#calculating total population for that year
TOTAL = race_ds['total_population'].sum()   #could have used "age_ds" but not "education_ds"
assert race_ds['total_population'].sum() == age_ds['total'].sum()
#print(TOTAL)
```

```
#print(dfa_race.loc[str(YEAR) + ':Q2'])
def net_worth(dataframe_dfa, dataframe_acs, year='2018', quater = 'Q1'):
  LIST = ['NAME', 'COUNTY', 'STATE', 'GEOID', 'total', 'total_population', 'totalAge25plus']
  columns = dataframe_acs.columns.to_list()
  for item in LIST:
    try:
      columns.remove(item)
      #print(item, "removed successfully")
    except:
      None
      #print(item, "was not available")
  df_dfa= dataframe_dfa.loc[year+ ':'+ quater]['Net worth']

  df = pd.DataFrame()

  for column in columns:
    temp = dataframe_acs[column].sum()
    assert dataframe_acs[column].isnull().sum() == 0
    net_worth_per_capita = df_dfa.loc[column]/temp    #in million
    df[column+'_networth_'+quater] = dataframe_acs[column]*net_worth_per_capita
  return df 
```

```
net_worth_race_q1 = net_worth(dfa_race, race_ds, YEAR, quater='Q1')
net_worth_race_q2 = net_worth(dfa_race, race_ds, YEAR, quater='Q2')
net_worth_race_q3 = net_worth(dfa_race, race_ds, YEAR, quater='Q3')
net_worth_race_q4 = net_worth(dfa_race, race_ds, YEAR, quater='Q4')
```

```
net_worth_age_q1 = net_worth(dfa_age, age_ds, YEAR, quater='Q1')
net_worth_age_q2 = net_worth(dfa_age, age_ds, YEAR, quater='Q2')
net_worth_age_q3 = net_worth(dfa_age, age_ds, YEAR, quater='Q3')
net_worth_age_q4 = net_worth(dfa_age, age_ds, YEAR, quater='Q4')
```

```
net_worth_education_q1 = net_worth(dfa_education, education_ds, YEAR, quater='Q1')
net_worth_education_q2 = net_worth(dfa_education, education_ds, YEAR, quater='Q2')
net_worth_education_q3 = net_worth(dfa_education, education_ds, YEAR, quater='Q3')
net_worth_education_q4 = net_worth(dfa_education, education_ds, YEAR, quater='Q4')
```

```
net_worth_df = pd.DataFrame()
net_worth_df = pd.concat([race_ds['total_population'], education_ds['totalAge25plus'],
                          net_worth_race_q1,net_worth_race_q2,net_worth_race_q3,net_worth_race_q4,
                          net_worth_age_q1,net_worth_age_q2,net_worth_age_q3,net_worth_age_q4,
                          net_worth_education_q1,net_worth_education_q2,net_worth_education_q3,net_worth_education_q4], axis = 1)
```
